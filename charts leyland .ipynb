```python
import networkx as nx
import plotly.graph_objects as go

# Define the graph structure
edges = [
    ("Data Fetching", "Feature Calculation"),
    ("Feature Calculation", "Asset Price Simulation"),
    ("GAN Generator", "GAN Training"),
    ("GAN Training", "GAN Discriminator")
]

# Create the graph object
G = nx.DiGraph()
G.add_edges_from(edges)

# Use a shell layout for hierarchical structure
shell_pos = nx.shell_layout(G)  # shell_layout places nodes in concentric circles

# Extract x and y positions for nodes and edges
edge_x = []
edge_y = []
for edge in G.edges():
    x0, y0 = shell_pos[edge[0]]
    x1, y1 = shell_pos[edge[1]]
    edge_x.append(x0)
    edge_x.append(x1)
    edge_y.append(y0)
    edge_y.append(y1)

node_x = []
node_y = []
for node in G.nodes():
    x, y = shell_pos[node]
    node_x.append(x)
    node_y.append(y)

# Create Plotly figure for network graph
fig = go.Figure()

# Plot edges (custom color for separation)
fig.add_trace(go.Scatter(x=edge_x, y=edge_y, line=dict(width=1, color='#888888'), hoverinfo='none', mode='lines'))

# Plot nodes (with custom style)
fig.add_trace(go.Scatter(x=node_x, y=node_y, mode='markers', hoverinfo='text',
                         marker=dict(color='skyblue', size=15, line=dict(width=2, color='black'))))

# Add labels for nodes
for i, node in enumerate(G.nodes()):
    fig.add_annotation(x=node_x[i], y=node_y[i], text=node, showarrow=False, font=dict(size=10, color="black"))

# Layout adjustments
fig.update_layout(
    title="Hierarchical Layout for Data Flow",
    showlegend=False,
    hovermode='closest',
    xaxis=dict(showgrid=False, zeroline=False),
    yaxis=dict(showgrid=False, zeroline=False),
    height=600,
    width=800
)

# Display the interactive plot
fig.show()
```
