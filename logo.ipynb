```python
from PIL import Image, ImageDraw, ImageFont

# Create a minimal logo
img_size = (256, 256)  # Image size
background_color = "white"
text_color = "black"
font_size = 100

# Initialize image
img = Image.new("RGB", img_size, background_color)
draw = ImageDraw.Draw(img)

# Load font
font = ImageFont.truetype("arial.ttf", font_size)

# Add text
text = "VA"
text_bbox = draw.textbbox((0, 0), text, font=font)  # Get bounding box of the text
text_width = text_bbox[2] - text_bbox[0]
text_height = text_bbox[3] - text_bbox[1]
text_position = ((img_size[0] - text_width) // 2, (img_size[1] - text_height) // 2)

draw.text(text_position, text, fill=text_color, font=font)

# Save the image
img.save("minimal_logo.png")
print("Logo saved as 'minimal_logo.png'")
```

```python
from PIL import Image, ImageDraw, ImageFont

# Create a fancier logo
img_size = (400, 400)  # Larger image for better detail
font_size = 120
text = "VA"

# Create a gradient background
img = Image.new("RGB", img_size, "white")
draw = ImageDraw.Draw(img)

# Gradient colors
color1 = (0, 102, 204)  # Dark blue
color2 = (255, 255, 255)  # White
for y in range(img_size[1]):
    r = color1[0] + (color2[0] - color1[0]) * y // img_size[1]
    g = color1[1] + (color2[1] - color1[1]) * y // img_size[1]
    b = color1[2] + (color2[2] - color1[2]) * y // img_size[1]
    draw.line([(0, y), (img_size[0], y)], fill=(r, g, b))

# Load a custom font
font_path = "/usr/share/fonts/truetype/dejavu/DejaVuSerif-Bold.ttf"  # Update this to a custom font if desired
font = ImageFont.truetype(font_path, font_size)

# Calculate text position
text_bbox = draw.textbbox((0, 0), text, font=font)
text_width = text_bbox[2] - text_bbox[0]
text_height = text_bbox[3] - text_bbox[1]
text_position = ((img_size[0] - text_width) // 2, (img_size[1] - text_height) // 2)

# Add text with shadow
shadow_offset = 4
draw.text((text_position[0] + shadow_offset, text_position[1] + shadow_offset), text, font=font, fill="black")  # Shadow
draw.text(text_position, text, font=font, fill="white")  # Main text

# Add a border around the image
border_width = 8
draw.rectangle(
    [(border_width // 2, border_width // 2), (img_size[0] - border_width // 2, img_size[1] - border_width // 2)],
    outline="black",
    width=border_width,
)

# Save the image
img.save("fancy_logo.png")
print("Fancy logo saved as 'fancy_logo.png'")
```
