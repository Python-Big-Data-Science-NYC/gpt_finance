```python
import tensorflow as tf
from tensorflow.keras import layers, models
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
from tensorflow.keras import backend as K
from sklearn.preprocessing import StandardScaler

# Step 1: Fetch data with a larger range for computational cost
ticker = '^TNX'  # U.S. 10-year Treasury yield (interest rates)
data = yf.download(ticker, start='2000-01-01', end='2024-01-01')  # Extended data range
interest_rates = data['Adj Close']

# Step 2: Normalize the interest rates
scaler = StandardScaler()
interest_rates_normalized = scaler.fit_transform(interest_rates.values.reshape(-1, 1))

# Step 3: Build the VAE model with increased complexity

latent_dim = 3  # Increased latent dimension for complexity

# Encoder model with more layers and neurons
inputs = layers.Input(shape=(1,))
x = layers.Dense(128, activation='relu')(inputs)  # Increased layer size
x = layers.Dense(64, activation='relu')(x)  # Increased layer size
x = layers.Dense(32, activation='relu')(x)
z_mean = layers.Dense(latent_dim, name='z_mean')(x)
z_log_var = layers.Dense(latent_dim, name='z_log_var')(x)

# Reparameterization trick (sampling from a normal distribution)
class Sampling(layers.Layer):
    def call(self, inputs):
        z_mean, z_log_var = inputs
        batch = K.shape(z_mean)[0]
        dim = K.int_shape(z_mean)[1]
        epsilon = K.random_normal(shape=(batch, dim))
        return z_mean + K.exp(0.5 * z_log_var) * epsilon

z = Sampling()([z_mean, z_log_var])

# Decoder model with more complexity
latent_inputs = layers.Input(shape=(latent_dim,))
x = layers.Dense(64, activation='relu')(latent_inputs)  # Increased layer size
x = layers.Dense(128, activation='relu')(x)  # Increased layer size
outputs = layers.Dense(1)(x)

# Instantiate the models
encoder = models.Model(inputs, [z_mean, z_log_var, z])
decoder = models.Model(latent_inputs, outputs)
vae_output = decoder(encoder(inputs)[2])

# Custom loss function as part of the Keras model
def vae_loss(inputs, vae_output, z_mean, z_log_var):
    xent_loss = K.mean(K.square(inputs - vae_output), axis=-1)
    kl_loss = - 0.5 * K.mean(1 + z_log_var - K.square(z_mean) - K.exp(z_log_var), axis=-1)
    return xent_loss + kl_loss

# Create a custom model that integrates the loss into the call method
class VAEWithLoss(models.Model):
    def __init__(self, encoder, decoder, **kwargs):
        super(VAEWithLoss, self).__init__(**kwargs)
        self.encoder = encoder
        self.decoder = decoder

    def call(self, inputs):
        z_mean, z_log_var, z = self.encoder(inputs)
        vae_output = self.decoder(z)
        loss = vae_loss(inputs, vae_output, z_mean, z_log_var)
        self.add_loss(loss)
        return vae_output

# Instantiate the VAE model with the loss integrated
vae = VAEWithLoss(encoder, decoder)
vae.compile(optimizer='adam')

# Step 4: Train the VAE with increased computational load
vae.fit(interest_rates_normalized, interest_rates_normalized, epochs=50, batch_size=16)  # More epochs and smaller batch size

# Step 5: Visualize the learned latent factors (Increased computation)
# The latent factors can be used as features for building credit risk models.
# For example, you can combine these factors with other features (e.g., borrower profiles)
# to predict the likelihood of default using a classifier such as Logistic Regression or XGBoost.
```

```python
# Step 6: Obtain predicted values
predicted_rates_normalized = vae.predict(interest_rates_normalized)

# Step 7: Rescale the normalized data back to the original scale
predicted_rates = scaler.inverse_transform(predicted_rates_normalized)
actual_rates = scaler.inverse_transform(interest_rates_normalized)

# Step 8: Plot actual vs predicted values
plt.figure(figsize=(12, 6))
plt.plot(data.index, actual_rates, label='Actual Interest Rates', color='blue', alpha=0.7)
plt.plot(data.index, predicted_rates, label='Predicted Interest Rates', color='red', alpha=0.7)
plt.title('Actual vs Predicted Interest Rates')
plt.xlabel('Date')
plt.ylabel('Interest Rate')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
from datetime import timedelta

# Step 9: Generate latent variables for prediction
latent_means, latent_log_vars, _ = encoder.predict(interest_rates_normalized)

# Sample new latent variables for prediction (for one year, assuming daily data)
num_days = 252  # Approximately 1 year of trading days
latent_samples = np.random.normal(size=(num_days, latent_dim))

# Decode the latent variables into predicted interest rates (normalized)
predicted_normalized_future_rates = decoder.predict(latent_samples)

# Rescale to the original scale
predicted_future_rates = scaler.inverse_transform(predicted_normalized_future_rates)

# Step 10: Generate future dates
last_date = data.index[-1]
future_dates = [last_date + timedelta(days=i) for i in range(1, num_days + 1)]

# Create a DataFrame for future predictions
future_predictions_df = pd.DataFrame({
    'Date': future_dates,
    'Predicted Interest Rate': predicted_future_rates.flatten()
})
future_predictions_df.set_index('Date', inplace=True)

# Step 11: Plot historical and predicted future interest rates
plt.figure(figsize=(14, 7))
plt.plot(data.index, scaler.inverse_transform(interest_rates_normalized), label='Historical Interest Rates', color='blue', alpha=0.7)
plt.plot(future_predictions_df.index, future_predictions_df['Predicted Interest Rate'], label='Predicted Future Interest Rates', color='green', linestyle='--')
plt.title('Historical and Predicted Future Interest Rates')
plt.xlabel('Date')
plt.ylabel('Interest Rate')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Step 9: Generate predictions for the next 5 years (252 trading days/year)
num_days = 252 * 5  # 5 years of trading days
num_samples = 100  # Number of samples to estimate mean and variation for each day

# Generate latent samples for each day and compute predictions
latent_means, latent_log_vars, _ = encoder.predict(interest_rates_normalized)
future_predictions = []

for _ in range(num_samples):
    latent_samples = np.random.normal(size=(num_days, latent_dim))  # Random latent variables
    predicted_normalized_future_rates = decoder.predict(latent_samples)  # Decode to predictions
    future_predictions.append(scaler.inverse_transform(predicted_normalized_future_rates).flatten())  # Rescale and flatten

# Convert to a NumPy array for easier manipulation
future_predictions = np.array(future_predictions)  # Shape: (num_samples, num_days)

# Compute the mean and standard deviation for each day
predicted_means = np.mean(future_predictions, axis=0)  # Mean for each day
predicted_stds = np.std(future_predictions, axis=0)    # Standard deviation for each day

# Step 10: Generate future dates
last_date = data.index[-1]
future_dates = [last_date + timedelta(days=i) for i in range(1, num_days + 1)]

# Create a DataFrame for future predictions
future_predictions_df = pd.DataFrame({
    'Date': future_dates,
    'Predicted Mean': predicted_means,
    'Predicted StdDev': predicted_stds
})
future_predictions_df.set_index('Date', inplace=True)

# Step 11: Plot historical data and predicted means with variation
plt.figure(figsize=(14, 7))

# Plot historical interest rates
plt.plot(data.index, scaler.inverse_transform(interest_rates_normalized), label='Historical Interest Rates', color='blue', alpha=0.7)

# Plot predicted mean
plt.plot(future_predictions_df.index, future_predictions_df['Predicted Mean'], label='Predicted Mean', color='green')

# Plot variation as a shaded region (mean ± std deviation)
plt.fill_between(
    future_predictions_df.index,
    future_predictions_df['Predicted Mean'] - future_predictions_df['Predicted StdDev'],
    future_predictions_df['Predicted Mean'] + future_predictions_df['Predicted StdDev'],
    color='green',
    alpha=0.2,
    label='Predicted StdDev Range'
)

plt.title('Historical and Predicted Interest Rates (Next 5 Years)')
plt.xlabel('Date')
plt.ylabel('Interest Rate')
plt.legend()
plt.grid(True)
plt.show()
```
