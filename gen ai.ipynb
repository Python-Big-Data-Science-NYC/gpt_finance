```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn3

# Define the sets for Investment, Risk, and Trading
investment = {'moodys2023rise', 'roosevelt2024risks', 'fsb2024ai', 'cognizant2024genai', 'mckinsey2024agents', 'aws2024making'}
risk = {'roosevelt2024risks', 'see2024credit', 'rethink2024risk', 'fsb2024ai', 'mckinsey2024agents'}
trading = {'fatemiFinVisionMultiAgentFramework2024', 'hanEnhancingInvestmentAnalysis2024', 'microfoundationsMultiagentMarketModel2024', 'zhangMultimodalFoundationAgent2024'}

# Create the Venn diagram
venn_labels = {
    '100': len(investment - (risk | trading)),  # Only in Investment
    '010': len(risk - (investment | trading)),  # Only in Risk
    '001': len(trading - (investment | risk)),  # Only in Trading
    '110': len(investment & risk - trading),  # Investment and Risk
    '101': len(investment & trading - risk),  # Investment and Trading
    '011': len(risk & trading - investment),  # Risk and Trading
    '111': len(investment & risk & trading)  # Investment, Risk, and Trading
}

# Plot the Venn diagram
plt.figure(figsize=(8, 6))
venn = venn3(subsets=venn_labels, set_labels=('Investment', 'Risk', 'Trading'))

# Show plot
plt.title("Venn Diagram of Investment, Risk, and Trading in Finance")



from datetime import datetime
timestamp = datetime.now().strftime('%Y%m%d_%H%M%S_%f')
plt.savefig(f"d1{timestamp}.png", dpi=300, bbox_inches='tight')


plt.show()
```
