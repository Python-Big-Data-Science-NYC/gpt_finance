```python
import matplotlib.pyplot as plt
import networkx as nx

# Define the steps and relationships
steps = {
    "Fetch Financial Data": ["Ticker: AAPL", "Period: 2022"],
    "Calculate Features": ["Returns", "Volatility", "ROC"],
    "Normalize Features": ["StandardScaler"],
    "Build VAE": ["Encoder", "Decoder"],
    "Train VAE": ["Loss: Reconstruction + KL"],
    "Latent Features": ["z_mean", "z_log_var", "z"],
    "Cox Model": ["Fit survival analysis"],
}

edges = [
    ("Fetch Financial Data", "Calculate Features"),
    ("Calculate Features", "Normalize Features"),
    ("Normalize Features", "Build VAE"),
    ("Build VAE", "Train VAE"),
    ("Train VAE", "Latent Features"),
    ("Latent Features", "Cox Model"),
]

# Create a directed graph
G = nx.DiGraph()

# Add nodes and edges
for step, details in steps.items():
    G.add_node(step, details="\n".join(details))
for edge in edges:
    G.add_edge(*edge)

# Plot the graph
plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G)

# Draw nodes
node_labels = {node: f"{node}\n{data['details']}" for node, data in G.nodes(data=True)}
nx.draw_networkx_nodes(G, pos, node_size=3000, node_color="lightblue", alpha=0.9)
nx.draw_networkx_labels(G, pos, labels=node_labels, font_size=10)

# Draw edges
nx.draw_networkx_edges(G, pos, arrowstyle="->", arrowsize=20, edge_color="gray")

plt.title("Pipeline Diagram: Financial Data to Survival Analysis", fontsize=14)
plt.axis("off")
from datetime import datetime
timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
filename = f"plot_{timestamp}.png"
plt.savefig(filename)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for the linear flow
steps = [
    "Fetch Financial Data",
    "Calculate Features",
    "Normalize Features",
    "Build VAE",
    "Train VAE",
    "Extract Latent Features",
    "Fit Cox Model"
]

plt.figure(figsize=(12, 4))
for i, step in enumerate(steps):
    plt.text(i, 0, step, ha='center', va='center', bbox=dict(boxstyle="round", facecolor="lightblue", alpha=0.8))

plt.plot(range(len(steps)), [0] * len(steps), marker='o', color='gray', markersize=10)
plt.ylim(-1, 1)
plt.axis("off")
plt.title("Linear Flow Diagram", fontsize=14)
plt.show()
```

```python
import matplotlib.pyplot as plt
import networkx as nx

# Define the tree structure
G = nx.DiGraph()
G.add_edges_from([
    ("Fetch Financial Data", "Returns"),
    ("Fetch Financial Data", "Volatility"),
    ("Fetch Financial Data", "ROC"),
    ("Returns", "VAE Training"),
    ("Volatility", "VAE Training"),
    ("ROC", "VAE Training"),
    ("VAE Training", "Latent Features"),
    ("Latent Features", "Cox Model")
])

# Draw the tree
plt.figure(figsize=(10, 8))
pos = nx.drawing.nx_agraph.graphviz_layout(G, prog="dot")
nx.draw(G, pos, with_labels=True, arrows=True, node_size=2000, node_color="lightgreen", font_size=10, alpha=0.9)
plt.title("Hierarchical Tree Diagram", fontsize=14)
from datetime import datetime
timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
filename = f"plot_{timestamp}.png"
plt.savefig(filename)
plt.show()
```

```python
import matplotlib.pyplot as plt
import networkx as nx

# Define the circular flow
steps = [
    "Fetch Financial Data",
    "Calculate Features",
    "Normalize Features",
    "Build VAE",
    "Train VAE",
    "Extract Latent Features",
    "Fit Cox Model"
]

# Create the directed graph
G = nx.DiGraph()
for i in range(len(steps)):
    G.add_edge(steps[i], steps[(i + 1) % len(steps)])  # Circular flow

# Draw the circular graph
plt.figure(figsize=(8, 8))
pos = nx.circular_layout(G)
nx.draw(G, pos, with_labels=True, node_size=3000, node_color="lightcoral", font_size=10, alpha=0.9)
plt.title("Circular Workflow Diagram", fontsize=14)
from datetime import datetime
timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
filename = f"plot_{timestamp}.png"
plt.savefig(filename)
plt.show()
```
