```python
import numpy as np
import matplotlib.pyplot as plt

# Define categories
categories = ['Scalability', 'Latency', 'Ease of Integration', 'Cost Efficiency', 'Reliability']
technologies = {
    'Kafka': [9, 7, 6, 8, 9],
    'Kinesis': [8, 8, 7, 6, 8],
    'Glue': [7, 6, 9, 7, 7]
}

# Convert labels to angles
num_vars = len(categories)
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]  # Complete the loop

# Plot radar chart
fig, ax = plt.subplots(figsize=(7, 7), subplot_kw=dict(polar=True))

for tech, values in technologies.items():
    values += values[:1]  # Repeat first value to close the chart
    ax.plot(angles, values, label=tech, linewidth=2)
    ax.fill(angles, values, alpha=0.2)

# Adjust aesthetics
ax.set_yticklabels([])
ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories)
plt.title('Comparison of Streaming & Integration Technologies')
plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))
plt.show()
```

```python
import matplotlib.pyplot as plt

# Define layers and positions
layers = ["Application Layer", "AI Processing Layer", "Data Integration Layer", "Streaming Layer", "Storage Layer"]
y_positions = range(len(layers), 0, -1)

fig, ax = plt.subplots(figsize=(6, 6))

# Draw rectangles for each layer
for i, layer in enumerate(layers):
    ax.barh(y_positions[i], width=6, height=0.6, align='center', color='skyblue', edgecolor='black')
    ax.text(0, y_positions[i], layer, fontsize=12, va='center', ha='center', weight='bold')

ax.set_xlim(-3, 3)
ax.set_yticks([])
ax.set_xticks([])
ax.set_title("Layered Architecture for AI Data Streaming")

plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Define nodes and edges
G = nx.DiGraph()
nodes = ["Data Sources", "Streaming (Kafka/Kinesis)", "ETL Integration (Glue)", "Vector DB (FAISS)", "AI Model"]
edges = [("Data Sources", "Streaming (Kafka/Kinesis)"), 
         ("Streaming (Kafka/Kinesis)", "ETL Integration (Glue)"), 
         ("ETL Integration (Glue)", "Vector DB (FAISS)"),
         ("Vector DB (FAISS)", "AI Model")]

G.add_nodes_from(nodes)
G.add_edges_from(edges)

# Position nodes
pos = nx.spring_layout(G, seed=42)

plt.figure(figsize=(8, 5))
nx.draw(G, pos, with_labels=True, node_size=3000, node_color="lightblue", font_size=10, edge_color="gray", arrows=True)
plt.title("AI Data Pipeline Flow")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Define timeline points
years = [2025, 2027, 2030, 2032, 2035]
events = [
    "Expansion of real-time AI pipelines (Kafka, Kinesis) \n[ApacheKafkaMission, ProcessingDataApache2017]",
    "Adoption of AI-driven data integration (AWS Glue, HybridDB) \n[AmazonDataIntegration, AnalyticDBPostgreSQLOnline]",
    "Advancements in retrieval-augmented generation (RAG) \n[DataStreamingGenerative]",
    "Vector databases achieving full AI-native search capabilities \n[RoleVectorDatabases2023]",
    "AI-driven autonomous decision-making systems"
]

# Create the timeline plot
fig, ax = plt.subplots(figsize=(8, 4))
ax.scatter(years, [1]*len(years), color="blue", s=100)

# Add event labels
for i, (year, event) in enumerate(zip(years, events)):
    ax.text(year, 1.02 if i % 2 == 0 else 0.98, event, ha='center', fontsize=10, rotation=30)

ax.set_yticks([])
ax.set_xticks(years)
ax.set_xlabel("Year")
ax.set_title("AI Data Streaming and Integration Roadmap (2025-2035)")
ax.grid(axis="x", linestyle="--", alpha=0.7)

plt.show()
```

```python
import matplotlib.pyplot as plt

# Define timeline points
years = [2025, 2027, 2030, 2033, 2035]
events = [
    "Expansion of real-time AI pipelines",
    "Adoption of AI-driven data integration",
    "Advancements in retrieval-augmented generation",
    "Vector databases achieving full AI-native search capabilities",
    "AI-driven autonomous decision-making systems"
]

# Create the timeline plot
fig, ax = plt.subplots(figsize=(8, 4))
ax.scatter(years, [1] * len(years), color="blue", s=100)

# Adjust vertical placement to minimize overlap
for i, (year, event) in enumerate(zip(years, events)):
    y_offset = 1.05 if i % 2 == 0 else 0.95
    ax.text(year, y_offset, event, ha='center', fontsize=10, rotation=20)

ax.set_yticks([])
ax.set_xticks(years)
ax.set_xlabel("Year")
ax.set_title("AI Data Streaming and Integration Roadmap (2025-2035)")
ax.grid(axis="x", linestyle="--", alpha=0.7)

plt.show()
```
