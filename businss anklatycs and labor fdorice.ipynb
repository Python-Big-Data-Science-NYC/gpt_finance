```python
import matplotlib.pyplot as plt
import pandas as pd
import bibtexparser
from collections import Counter

# Load the .bib file
with open("/home/j/gitrepos/ieee_conf_genai_fin_risk/business_analytics_genai/businesses_genai_no_fin.bib", encoding="utf-8") as bibtex_file:
    bib_database = bibtexparser.load(bibtex_file)

# Extract year and title information from the .bib file
titles = [entry.get("title", "") for entry in bib_database.entries]
years = [int(entry.get("year", "2025")) for entry in bib_database.entries]  # Default to 2025 if no year

# Step 1: Manually define or extract categories based on titles (or use keywords from titles)
categories = []
for title in titles:
    if "AI" in title or "artificial" in title:
        categories.append("AI")
    elif "labor" in title or "work" in title:
        categories.append("Labor Market")
    elif "finance" in title:
        categories.append("Finance")
    elif "bank" in title or "central" in title:
        categories.append("Finance")
    else:
        categories.append("Other")

# Step 2: Calculate significance based on frequency of occurrence in titles (for simplicity, assume higher frequency means higher significance)
category_counts = Counter(categories)
max_count = max(category_counts.values())
significance = [category_counts[category] / max_count for category in categories]  # Scale significance

# Step 3: Create DataFrame with extracted data
df = pd.DataFrame({
    "Year": years,
    "Event": titles,
    "Category": categories,
    "Significance": significance,
})

# Step 4: Map categories to numeric values for the y-axis
category_to_numeric = {
    "AI": 1,
    "Labor Market": 2,
    "Finance": 3,
    "Other": 4,
}
df["Category_Numeric"] = df["Category"].map(category_to_numeric)

# Step 5: Map categories to colors
category_colors = {
    "AI": "blue",
    "Labor Market": "green",
    "Finance": "orange",
    "Other": "red",
}

# Step 6: Create the bubble chart
plt.figure(figsize=(12, 8))

# Plot each category separately
for category, color in category_colors.items():
    subset = df[df["Category"] == category]
    plt.scatter(
        subset["Year"],
        subset["Category_Numeric"],
        s=subset["Significance"] * 1000,  # Scale bubble size
        c=color,
        label=category,
        alpha=0.6,
        edgecolors="black",
    )

# Add labels and title
plt.title("Future Predictions from AI Research Papers (20252035)", fontsize=16)
plt.xlabel("Year", fontsize=14)
plt.ylabel("Category", fontsize=14)
plt.xticks(range(2025, 2036, 1), rotation=45)  # Set x-axis ticks for each year
plt.yticks(ticks=[1, 2, 3, 4], labels=category_to_numeric.keys(), fontsize=12)  # Map numeric y-values back to category names
plt.grid(True, linestyle="--", alpha=0.6)

# Add legend
plt.legend(title="Category", bbox_to_anchor=(1.05, 1), loc="upper left")

# Add event labels
for i, row in df.iterrows():
    plt.text(
        row["Year"],
        row["Category_Numeric"],
        row["Event"],
        fontsize=9,
        ha="left",
        va="bottom",
        wrap=True,
    )

# Show the plot
plt.tight_layout()
plt.show()
```
