```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the radar chart
labels = np.array(["Reskilling", "Unemployment", "ChatGPT", "Future Jobs", "Economy"])
values = np.array([3, 2, 6, 4, 5])  # Numbers from the analysis

# Number of variables
num_vars = len(labels)

# Compute angle of each axis
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

# The plot is circular, so we need to "complete the loop" and append the start to the end
values = np.concatenate((values, [values[0]]))
angles += angles[:1]

# Initialize the radar chart
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))

# Draw one axis per variable and add labels
ax.set_theta_offset(np.pi / 2)
ax.set_theta_direction(-1)
ax.set_rlabel_position(0)
plt.xticks(angles[:-1], labels)

# Draw ylabels
plt.yticks([1, 2, 3, 4, 5, 6], ["1", "2", "3", "4", "5", "6"], color="grey", size=7)
plt.ylim(0, 6)

# Plot data
ax.plot(angles, values, linewidth=2, linestyle='solid', label='Keyword Frequency')
ax.fill(angles, values, 'b', alpha=0.1)

# Add a legend
plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))

# Add a title
plt.title("Keyword Frequency in Generative AI Research", size=14, y=1.1)
plt.savefig("radar_chart.png", dpi=300, bbox_inches='tight')  # Save as PNG
# Show the plot
plt.show()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Word frequencies from the analysis
word_frequencies = {
    "Artificial": 12,
    "Intelligence": 10,
    "Education": 8,
    "Generative": 7,
    "AI": 6,
    "ChatGPT": 6,
    "Work": 5,
    "Future": 5,
    "Learning": 4,
    "Students": 4,
    "Impact": 4,
    "Jobs": 4,
    "Economy": 3,
    "Reskilling": 3,
    "Teaching": 3
}

# Generate the word cloud
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color='white',
    colormap='viridis',  # Color scheme
    max_words=50,        # Limit to top 50 words
    relative_scaling=0.5  # Adjust word sizes
).generate_from_frequencies(word_frequencies)

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # Remove axes
plt.title("Word Cloud of Generative AI Research Topics", size=16, pad=20)
plt.savefig("word_cloud.png", dpi=300, bbox_inches='tight')  # Save as PNG
plt.show()
```
